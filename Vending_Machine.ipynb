```python
#USING THE IMPORT FUNCTION
import time
print("\033[1;35;40m WELCOME TO OUR VENDING MACHINE")  # addition of  PURPLE  color with black background to it.
#input the amount of money..
number_of_AED = int(input("How many  dirhams would you like to insert? "))

#The money that will be  inserted by the customer using..
change = round(number_of_AED)

print ("\nYou have entered AED", change)
time.sleep(2)
# The below are the items along with their specific prices.. 

item_1 = "Water"
price_1 = 1.5
item_2 = "KITKAT"
price_2 = 3.75
item_3 = "Apple juice"
price_3 = 3.20
item_4 = "Cup cake"
price_4 = 2.25
item_5 = "Lays Chips"
price_5 = 2.00
item_6 = "Peanuts"
price_6 = 2.50
 
#The items purchased..
Water_purchased = 0
KITKAT_purchased = 0
Apple_juice_purchased = 0
Cup_cake_purchased = 0
Lays_Chips_purchased = 0
Peanuts_purchased = 0

print ("There are 6 items available to pick from;\n")
time.sleep(2)
print (" Item                   Prices")
print (" ----                   -----")
print (" {}         ----->  {} ".format(item_1, price_1))
print (" {}  ----->  {} ".format(item_2, price_2))
print (" {}     ----->  {} ".format(item_3, price_3))
print (" {}       ----->  {} ".format(item_4, price_4))
print (" {}    ----->  {} ".format(item_5, price_5))
print (" {}          ----->  {} ".format(item_6, price_6))
print ("")

while change > 0:
    customers_selection = input("What would you like to buy? Type N when you are finished \n")
    if customers_selection == "Water" or customers_selection == "Water" and change >= price_1:
        print ("You have selected a", item_1, "these cost", price_1, "each,")
        change = round((change - price_1),2)
        Water_purchased = (Water_purchased + 1)
        
        print ("Your balance remaining : AED", change)


    elif customers_selection == "KITKAT" or customers_selection == "KITKAT" and change >= price_2:
        print ("You have selected a", item_2, "these cost", price_2, "each,")
        change = round((change - price_2),2)
        KITKAT_purchased = (KITKAT_purchased + 1)
      
        print ("Your balance remaining : AED", change)


    elif customers_selection == "Apple juice" or customers_selection == "Apple juice" and change >= price_3:
        print ("You have selected a", item_3, "these cost", price_3, "each,")
        change = round((change - price_3),2)
        Apple_juice_purchased = (Apple_juice_purchased + 1)
        
        print ("Your balance remaining : AED", change)


    elif customers_selection == "Cup cake" or customers_selection == "Cup cake" and change >= price_4:
        print ("You have selected a", item_4, "these cost", price_4, "each,")
        change = round((change - price_4),2)
        Cup_cake_purchased = (Cup_cake_purchased + 1)
        
        print ("Your balance remaining ", change)


    elif customers_selection == "Lays Chips" or customers_selection == "Lays Chips" and change >= price_5:
        print ("You have selected a", item_5, "these cost", price_5, "each,")
        change = round((change - price_5),2)
        Lays_Chips_purchased = (Lays_Chips_purchased + 1)
        
        print ("Your balance remaining ", change)


    elif customers_selection == "Peanuts" or customers_selection == "Peanuts" and change >= price_6:
        print ("You have selected a", item_6, "these cost", price_6, "each,")
        change = round((change - price_6),2)
        Peanuts_purchased = (Peanuts_purchased + 1)
        
        print ("Your balance remaining ", change)


    elif customers_selection == "N" or customers_selection == "n":
       
        break

    elif change <= 0:
        print ("You have insufficient balance.")
        
        break
    else:
        print ("There has been an error or you may not have enough credit.")
    

        
print ("\nYou have AED", change, "remaining.")
time.sleep(2)
#This will be the receipt of customer..
print ("\nHere is your receipt;")
time.sleep(2)
if Water_purchased > 0:
    print ("You purchased", Water_purchased, item_1, "at AED", (price_1 * Water_purchased))
if KITKAT_purchased > 0:
    print ("You purchased", KITKAT_purchased, item_2, "at AED", (price_2 * KITKAT_purchased))
if Apple_juice_purchased > 0:
    print ("You purchased", Apple_juice_purchased, item_3, "at AED", (price_3 * Apple_juice_purchased))
if Cup_cake_purchased > 0:
    print ("You purchased", Cup_cake_purchased, item_4, "at AED", (price_4 * Cup_cake_purchased))
if Lays_Chips_purchased > 0:
    print ("You purchased", Lays_Chips_purchased, item_5, "at AED", (price_5 * Lays_Chips_purchased))
if Peanuts_purchased > 0:
    print ("You purchased", Peanuts_purchased, item_6, "at AED", (price_6 * Peanuts_purchased))
print("\033[1;35;40m")       # addition of  PURPLE  color with black background to it.
print("\033[1;35;40m THANK YOU FOR USING THE VENDING MACHINE")
```
